### Clase Asignatura
### 📚 Descripción del Problema: Clase Asignatura
El objetivo es modelar un sistema para la administración de unidades curriculares dentro de una institución educativa. Cada asignatura se define como una entidad independiente con carga académica y una identificación única.

1. Definición de Atributos
Los atributos son los datos que caracterizan a cada unidad de aprendizaje:

* nombre: Cadena de texto que describe la materia (ej. "Cálculo Integral").

* codigo: Identificador alfanumérico único para control administrativo (ej. "MAT102").

* creditos: Valor numérico que representa el peso académico o las horas de dedicación requeridas.

2. Definición de Funciones (Métodos)
Los métodos representan el comportamiento o las operaciones que podemos realizar con la asignatura:

* leerDatos(): Captura la información necesaria desde el teclado o una base de datos para configurar la materia.

* mostrarDatos(): Imprime en pantalla el resumen detallado de la asignatura (su "ficha técnica").

3. Instanciación de Objetos
Aquí es donde el molde se convierte en algo concreto. Creamos dos instancias que ocupan espacios distintos en la memoria:

* clase1: Objeto destinado a almacenar la información de la primera materia capturada.

* clase2: Objeto independiente que permite gestionar una segunda materia de forma simultánea.

![texto del vínculo](infoexaIIclassxxx.png.png)

```python
class Asignatura:
    def __init__(self):
        # Atributos inicializados vacíos
        self.nombre = ""
        self.codigo = ""
        self.creditos = 0

    def leerDatos(self):
        print(f"\n--- 📚 Registro de Nueva Asignatura ---")
        self.nombre = input("Nombre de la asignatura: ")
        self.codigo = input("Código administrativo: ")
        self.creditos = input("Número de créditos: ")

    def mostrarDatos(self):
        print("\n-------------------------------")
        print(f"📖 ASIGNATURA: {self.nombre.upper()}")
        print(f"🔑 Código: {self.codigo}")
        print(f"🎓 Créditos: {self.creditos}")
        print("-------------------------------")

# --- FLUJO INTERACTIVO CON 2 OBJETOS ---

# Creación del objeto clase1
clase1 = Asignatura()
clase1.leerDatos()

# Creación del objeto clase2
clase2 = Asignatura()
clase2.leerDatos()

# Mostrar los resultados finales
print("\n📋 RESUMEN DEL PLAN DE ESTUDIOS")
clase1.mostrarDatos()
clase2.mostrarDatos()
```
